### Respuesta.

El aprendizaje por refuerzo es una **rama del aprendizaje automático** en la que un agente autónomo aprende a tomar decisiones secuenciales mediante la interacción con un entorno. A diferencia del aprendizaje supervisado, no conoce de primeras el resultado de la información que se le pasa, sino que el agente recibe señales en forma de **recompensas o penalizaciones**, lo que le permite aprender qué acciones conducen a mejores resultados a largo plazo. La principal característica de cada rama es:

  * El aprendizaje supervisado se basa en datos etiquetados.

  * El no supervisado busca estructuras en datos no etiquetados.

  * El aprendizaje por refuerzo, por su parte, se basa en la interacción con un entorno dinámico, donde el agente aprende a través de pruebas y errores.

Los dos componentes principales de este tipo de aprendizaje son:

  1. El agente que toma decisiones y aprende a actuar.

  2. El entorno con el que el agente interactúa, generando estados y recompensas en función de los resultados obtenidos por las acciones del agente.

Respecto al ciclo de interacción podríamos decir que se basa en cinco pasos:

  1. El agente observa el estado actual del entorno.

  2. Toma una acción según su política actual.

  3. El entorno responde con una nueva observación (estado siguiente) y una recompensa.

  4. El agente actualiza su conocimiento con esta experiencia.

  5. El proceso se repite hasta alcanzar una condición de parada (como un número de episodios o una tarea resuelta).

Pasando a definir cada elemento, podemos decir que el agente es una entidad que decide que acción realizar en función de lo que observa, y el entorno es aquello que es observado por el agente, este es dinámico pues las acciones provocan un cambio en él, pudiendo ser este positivo o negativo en función del objetivo del agente.

Pasando a describir un poco más los elementos sobre los que se trabajan en esta rama:

  * Un estado es una representación del entorno en un momento dado. El espacio de estados incluye todos los posibles estados que puede tomar el entorno.

  * Una acción (o como hemos visto en la carrera, aristas) es una decisión que el agente puede tomar. El espacio de acciones agrupa todas las acciones disponibles en cada estado.

  * Es un valor numérico que indica el beneficio (positivo o negativo) obtenido por una acción. Sirve como métrica para que el agente aprenda. Para facilitar el trabajo realizado por este tipo de algoritmos los valores sobre los que se trabaja se transformarán al rango [0,1]. Esto es así porque los distintos datasets suelen tener rangos diferentes (256 si hablamos de píxeles, 5-10 si hablamos de valoraciones), por lo que es fundamental unificar esta métrica.

  * La política es la estrategia que sigue el agente para decidir qué acción tomar en cada estado. Dentro de la política podemos separarla en:
    * **Determinista**: Siempre elige la misma acción para un estado. Un ejemplo de esto sería el agente de la aspiradora, si el suelo está manchado, se tomará la opción de limpiar seguro.
    * **Estocástica**: Elige una acción según una distribución de probabilidad. Un ejemplo de esto podría ser un juego de mesa como el parchís, en el que hay que tener en cuenta diferentes sucesos que pueden ocurrir para determinar cuál es la opción más correcta.

  * Una experiencia podríamos definirla como un `struct` formado por: estado, acción, recompensa y siguiente estado, en el ámbito del aprendizaje por refuerzo esto se considera un nodo, mientras que en la IA simbólica este concepto de nodo sería el equivalente al (estado) definido previamente.

  * Episodio (o plan) es la secuencia completa de acciones que lleva desde un estado inical hasta un estado terminal. La trayectoria puede referirse a una secuencia (incompleta o total) de estados, acciones y recompensas, que forma parte de un episodio, la diferencia entre ambos es que un episodio siempre termina (como sería ganar al parchís), mientras que la trayectoria puede ser parcial (comer una ficha enemiga).

El entorno en aprendizaje por refuerzo suele verse como un Proceso de Decisión de Markov (MDP), que tiene como base que el siguiente estado depende únicamente del estado actual y de la acción tomada (propiedad de Markov).

Volviendo al ejemplo de la aspiradora:
  * **Estados**: {izquierda limpio, izquierda sucio, derecha limpio, derecha sucio}

  * **Acciones**: {ir a la izquierda, ir a la derecha, limpiar}

  * Las transiciones dependen solo del **estado actual** y de la **acción** (no del historial).

  * El agente recibirá una recompensa positiva si limpia la posición actual y esta está manchada.

El objetivo del agente es maximizar la recompensa acumulada a lo largo del tiempo, por esto aparece el concepto de retorno que es la suma de las recompensas obtenidas a partir de cierto momento, este valor puede ser: finito si el episodio termina o infinito si se usa un factor de descuento. En lo que respecta al factor de descuento, este se encuentra en el intervalo [0,1] y su objetivo es reducir el valor de las recompensas futuras, esto se usa para: favorecer decisiones que dan beneficios a corto plazo (en este ejemplo sería un valor bajo) o valorar el largo plazo (en el que el valor sería alto).

El valor de un estado es la expectativa del retorno que puede conseguir el agente desde un cierto estado 'x' siguiendo una política, a mayor valor de x mejor es el resultado esperado. Esto tiene similitud con el concepto f(x) visto en la heurística de A*.

Es la expectativa del retorno al tomar la acción a en el estado s y luego seguir una política. Informa al agente de qué acción le conviene más en cada situación. Esto tiene similitud con la predicción que se hace en A* de h(x)

### IMPORTAR LIBRERÍAS Y CREAR ENTORNO.

````python
"""
=== ALGORITMO Q-LEARNING: DESCRIPCIÓN DEL FUNCIONAMIENTO ===

Q-Learning es un algoritmo de aprendizaje por refuerzo off-policy que busca aprender
la función de valor acción-estado óptima Q*(s,a) sin necesidad de conocer el modelo
del entorno.

PASOS DEL ALGORITMO:
1. INICIALIZACIÓN:
   - Crear Q-table con valores iniciales (normalmente ceros)
   - Definir hiperparámetros: α (tasa aprendizaje), γ (factor descuento), ε (exploración)

2. PARA CADA EPISODIO:
   a) Reiniciar entorno y obtener estado inicial s
   b) PARA CADA PASO DEL EPISODIO:
      - Seleccionar acción a usando política ε-greedy:
        * Con probabilidad ε: acción aleatoria (EXPLORACIÓN)
        * Con probabilidad 1-ε: acción con mayor Q(s,a) (EXPLOTACIÓN)
      - Ejecutar acción a en el entorno
      - Observar recompensa r y nuevo estado s'
      - ACTUALIZAR Q-VALUE usando ecuación de Bellman:
        Q(s,a) ← Q(s,a) + α[r + γ max_a' Q(s',a') - Q(s,a)]
      - Actualizar estado: s ← s'
      - Si estado terminal: terminar episodio
   c) Decrementar ε (reducir exploración gradualmente)

3. CONVERGENCIA: Repetir hasta que Q-values se estabilicen

FUNCIÓN DE ACTUALIZACIÓN EXPLICADA:
- Q(s,a): Valor actual estimado para el par estado-acción
- α: Tasa de aprendizaje [0,1] - controla cuánto se actualiza el Q-value
- r: Recompensa inmediata observada
- γ: Factor de descuento [0,1] - importancia de recompensas futuras
- max_a' Q(s',a'): Mejor Q-value posible desde el siguiente estado s'
- [r + γ max_a' Q(s',a') - Q(s,a)]: Error TD (diferencia temporal)

El algoritmo es OFF-POLICY porque aprende Q* independientemente de la política
seguida durante la exploración (ε-greedy).
"""

import gymnasium as gym
from gymnasium.wrappers import TimeLimit
import numpy as np
import matplotlib.pyplot as plt
````

### INSTANCIAR EL ENTORNO CON TIME LIMIT

````python
# Crear entorno CliffWalking con límite de 500 pasos
env = TimeLimit(gym.make("CliffWalking-v0"), max_episode_steps=500)

# Ver info básica del entorno
print("Espacio de observación:", env.observation_space)
print("Espacio de acciones:", env.action_space)
````

### PROBAR reset() Y test()

````python
state, info = env.reset()
print("Estado inicial:", state)

# Ejecutar una acción de prueba (por ejemplo, Derecha = 1)
new_state, reward, terminated, truncated, info = env.step(1)
print("Nuevo estado:", new_state)
print("Recompensa:", reward)
print("Terminado:", terminated)
print("Truncado:", truncated)
````

### DEFINIR POLÍTICAS

````python
n_episodes = 100
n_actions = env.action_space.n

# Política aleatoria
def random_policy(state):
    return np.random.choice(n_actions)

# Política manual (naiva)
def naive_optimal_policy(state):
    if state % 12 < 11:  # no está en la última columna
        return 1  # derecha
    else:
        return 2  # abajo
````

### FUNCIÓN PARA EJECUTAR POLÍTICAS

````python
def run_policy(policy_fn):
    rewards = []
    for _ in range(n_episodes):
        state, _ = env.reset()
        total_reward = 0
        done = False
        while not done:
            action = policy_fn(state)
            state, reward, terminated, truncated, _ = env.step(action)
            total_reward += reward
            done = terminated or truncated
        rewards.append(total_reward)
    return rewards
````

### COMPARACIÓN TRAS LA EJECUCIÓN

````python
random_rewards = run_policy(random_policy)
naive_rewards = run_policy(naive_optimal_policy)

# Medias
print(f"Media recompensa aleatoria: {np.mean(random_rewards)}")
print(f"Media recompensa naiva: {np.mean(naive_rewards)}")

# Graficar resultados
plt.plot(random_rewards, label="Aleatoria")
plt.plot(naive_rewards, label="Naiva/Óptima")
plt.xlabel("Episodio")
plt.ylabel("Recompensa total")
plt.title("Comparación de políticas en CliffWalking")
plt.legend()
plt.grid(True)
plt.show()

env.close()
````

Los algoritmos clásicos de Aprendizaje por Refuerzo se suelen modelar como un Proceso de Decisión de Markov (MDP), que está formado por los siguientes componentes:
1. **Estados (S)**: reúne todas las posibles situaciones en las que el agente puede encontrarse en el entorno sobre el que se está trabajando.
2. **Acciones (A)**: todas las posibles acciones que el agente puede realizar desde un estado.
3. **Probabilidad de Transición (P(s’ | s,a))**: esto indica la probabilidad de transitar del estado s al siguiente estado s’ tras realizar la acción determinada por a.
4. **Recompensa (R(s,a) o R(s, a, s’))**: recompensa obtenida por el agente al realizar la acción a desde el estado s.
5. **Política (π(s) o π(a|s))**: indica la estrategia que define qué acción se debe tomar en cada estado en el que se encuentra el agente. Al realizar ejercicios de aprendizaje por refuerzo buscamos encontrar una política óptima que maximice la recompensa acumulada.
6. **Factor de descuento (γ)**: es un valor en el rango 0-1 que marca la importancia de las recompensas futuras en comparación con las recompensas inmediatas. Cuanto mayor sea el valor de γ mayor peso tendrán las recompensas futuras.
7. **Función de Valor (V(s))**: muestra la recompensa que se espera obtener a largo plazo a partir del estado s.
8. **Q-función (Q(s,a))**: representa la recompensa esperada a largo plazo de tomar una acción a en un estado s y luego seguir una política óptima buscada previamente.

Nota: la diferencia entre V(s) y Q(s,a) es que la primera responde a la pregunta ¿Cómo de bueno es este estado?, mientras que la segunda responde a ¿Cómo de bueno es realizar esta acción en este estado?

**Algoritmos pre-Deep Learning**

Antes de que llegase el Deep Learning exístian varios algoritmos que abordaban los problemas de aprendizaje por refuerzo, estos se clasifican en métodos basados en valores o en políticas, además de si son “on-policy” o “off-policy” (la diferencia entre ambos es que en on-policy el agente aprende la política que está usando para explorar el entorno mientras que en off-policy el agente aprende la política óptima independientemente de la política que está usando para explorar el entorno).

Los **métodos basados en valores** buscan aprender una función de valor (V(s) o Q(s,a)) que índique la “bondad” (entendemos bondad como el retorno total esperado a largo plazo) de los estados o pares estado-acción. Tras aprender la función de valor, la política se puede derivar seleccionado la acción que mayor valor aporte. Entre estos métodos encontramos:
1. **Value Iteration**: es un algoritmo de programación dinámica que calcula la función de valor óptima V*(s)* iterativamente. Se basa en la Ecuación de Bellman de Optimalidad y va actualizando continuamente el valor de cada estado hasta que converge. Tras converger, la política óptima π*(s) se puede derivar seleccionando la acción que maximiza la recompensa esperada.
2. **Q-Learning**: es un algoritmo de aprendizaje por refuerzo sin modelo y off-policy que aprende una función de acción-valor óptima Q(s,a) directamente. Este algoritmo mantiene una tabla llamada Q-table que mapea los pares estado-acción a sus valores esperados.

Los **métodos basados en políticas** buscan aprender directamente una política óptima sin necesidad de calcular una función de valor, el principal algoritmo es:
1. **Policy Iteration**: este algoritmo alterna entre dos fases: evaluación de política y mejora de política. En la primera, se calcula la función de valor para la política actual, mientras que en la segunda la política se actualiza de forma Greedy con respecto a la función de valor recién evaluada. Se va repitiendo este proceso hasta que la política converge.

Los **métodos de diferencia temporal** aprenden directamente de la experiencia sin un modelo del entorno, esto lo hacen utilizando la diferencia entre estimaciones de valor en pasos de tiempo sucesivos para actualizar las funciones de valor. El principal algoritmo es:
1. **SARSA (State-Action-Reward-State-Action)**: como se ha dicho previamente, es un algoritmo sin modelo y on-policy. Cabe aclarar que el nombre SARSA viene de la tupla (S, A, R, S’, A’) que se utiliza para actualizar la función Q.

Hemos seleccionado **Q-Learning** debido a su popularidad y su naturaleza “off-policy”, que nos parece más interesante que “on-policy”.

Q-Learning es un algoritmo de aprendizaje por refuerzo **sin modelo y off-policy** cuyo objetivo es encontrar la función de acción-valor óptima Q*(s,a) que representa la máxima recompensa esperada que se puede obtener al tomar la acción a desde el estado s y posteriormente seguir la política óptima definida.

Este algoritmo mantiene una tabla **Q-Table** que guarda los valores Q para cada par estado-acción. Inicialmente estos valores suelen ser 0 o valores pequeños, y conforme el agente va interactuando con el entorno, actualiza estos valores basándose en las recompensas recibidas.

**Pasos del Algoritmo Q-Learning**

El proceso de Q-Learning se lleva a cabo en episodios, donde cada episodio termina cuando el agente alcanza un estado terminal o un número máximo de pasos.

**Inicialización**
* Inicializar la Q-Table con ceros o valores aleatorios pequeños para todos los pares estado-acción (s, a).
* Definición de los hiperparámetros:
  1. Tasa de aprendizaje (𝛂): Un valor entre 0 y 1 que determina cuánto se actualizan los valores Q en cada paso. Un 𝛂 alto significa que la nueva información sobreescribe la antigua más rápidamente.
  2. Factor de descuento (ϒ): Un valor entre 0 y 1 que sopesa la importancia de las recompensas futuras.
  3. ϵ (épsilon): Para la política de exploración (generalmente ϵ-greedy). Determina la probabilidad de que el agente elija una acción aleatoria (exploración) en lugar de la acción con el Q-value más alto (explotación). ϵ suele decrecer con el tiempo.
  4. Número de episodios de entrenamiento.

**Bucle para cada episodio**:
* Establecer el estado inicial del entorno s.
* Bucle para cada paso del episodio (hasta alcanzar un estado terminal o el máximo de pasos):
* Seleccionar una acción (a): El agente selecciona una acción a para el estado actual s utilizando una política de exploración, comúnmente la política ϵ-greedy.
          Con probabilidad ϵ, el agente elige una acción aleatoria.
          Con probabilidad 1-ϵ, el agente elige la acción a que maximiza Q(s, a) en el estado actual s (explotación).
* Ejecutar la acción (a): El agente realiza la acción a en el entorno.

* Observar la retroalimentación: El entorno transita a un nuevo estado s’ y el agente recibe inmediatamente una recompensa r.

* Actualizar el valor Q: El Q-value para el par (s, a) se actualiza utilizando la siguiente función de actualización:

      Q(s, a) ← Q(s, a) + ɑ + [r + ϒ a’ max Q(s’, a’) - Q(s, a)]
Donde:

      Q(s, a): Es el Q-value actual del par estado-acción (s, a).
      ɑ: Es la tasa de aprendizaje. Determina la magnitud del ajuste del Q-value. Un ɑ alto hace que el agente se adapte rápidamente a la nueva información, mientras que un ɑ bajo suaviza los cambios.
      r: Es la recompensa inmediata recibida al tomar la acción a en el estado s y transitar a s’.
      ϒ: Es el factor de descuento. Indica la importancia e las recompensas futuras. Un ϒ cercano a 0 hace que el agente se enfoque solo en recompensas inmediatas, mientras que un ϒ cercano a 1 considera más las recompensas futuras.
      max a’ Q(s’, a’): Es el máximo Q-value posible para el siguiente estado s’ entre todas las acciones posibles a’. Este término representa la recompensa futura esperada que el agente podría obtener si actuara óptimamente desde el nuevo estado s’.
      Q(s, a) (dentro del corchete): Es la estimación anterior del Q-value para el par (s, a). La expresión r + ϒ max a’ Q(s’, a’) - Q(s, a) se conoce como el error de diferencia temporal (TD Error). Representa la diferencia entre la recompensa esperada del futuro (recompensa inmediata más recompensa futura descontada) y la estimación actual del Q-value.

* Actualizar el estado actual: s ← s’.

**Convergencia**

El proceso de aprendizaje continúa hasta que los valores Q se estabilizan (convergen a los valores Q óptimos) o se alcanza un número predefinido de episodios.

**Significado de los Términos en la Fundación de Actualización**

La ecuación de actualización de Q-Learning es una aplicación de la Ecuación de Bellman de Optimalidad en un contexto de aprendizaje. Busca mejorar la estimación actual de Q(s, a) moviéndola hacia un valor “más correcto” basado en la experiencia reciente.
* Q(s, a) ← Q(s, a) + ɑ x (Nuevo Valor Estimado-Valor Actual)
      El término ɑ x representa el “ajuste” o la “corrección” que se aplica al Q-value actual.
* Recompensa Inmediata: Es lo que el agente obtiene instantáneamente por su acción. Es un componente directo de la “bondad” de la transición.

* Recompensa Futura Descontada (ϒ max a’ Q(s’, a’)): Es la estimación de la recompensa máxima que el agente puede esperar obtener del siguiente estado s’ en adelante, descontada por ϒ. El uso de max a’ Q(s’, a’) es lo que hace que Q-Learning sea off-policy, ya que aprende el valor de la acción óptima en el siguiente estado, incluso si la acción realmente tomada para llegar a s’ no fue la óptima según la política actual.

* Estimación Actual (Q(s, a)): Es la creencia actual del agente sobre el valor de tomar la acción a en el estado s.

* Error TD: La diferencia entre el nuevo valor esperado (recompensa inmediata + recompensa futura descontada) y la estimación actual. Este error impulsa el aprendizaje; si el error es grande, el ajuste del Q-Value será mayor.

**En resumen**, Q-Learning permite al agente aprender la “calidad” de tomar una acción en un estado dado, incluso si el agente no siempre sigue la política óptima durante la fase de entrenamiento. Esto lo hace muy flexible y ha sido una base para muchos avances posteriores en el aprendizaje por refuerzo.

### Descripción del Algoritmo Q-Learning

**=== ALGORITMO Q-LEARNING: DESCRIPCIÓN DEL FUNCIONAMIENTO ===**

Q-Learning es un algoritmo de aprendizaje por refuerzo off-policy que busca aprender la función de valor acción-estado óptima Q*(s,a) sin necesidad de conocer el modelo del entorno.

**PASOS DEL ALGORITMO:**

1. **INICIALIZACIÓN:**
   - Crear Q-table con valores iniciales (normalmente ceros)
   - Definir hiperparámetros: α (tasa aprendizaje), γ (factor descuento), ε (exploración)

2. **PARA CADA EPISODIO:**
   a) Reiniciar entorno y obtener estado inicial s
   
   b) **PARA CADA PASO DEL EPISODIO:**
      - Seleccionar acción a usando política ε-greedy:
        * Con probabilidad ε: acción aleatoria (**EXPLORACIÓN**)
        * Con probabilidad 1-ε: acción con mayor Q(s,a) (**EXPLOTACIÓN**)
      - Ejecutar acción a en el entorno
      - Observar recompensa r y nuevo estado s'
      - **ACTUALIZAR Q-VALUE** usando ecuación de Bellman:
        
        **Q(s,a) ← Q(s,a) + α[r + γ max_a' Q(s',a') - Q(s,a)]**
      - Actualizar estado: s ← s'
      - Si estado terminal: terminar episodio
   
   c) Decrementar ε (reducir exploración gradualmente)

3. **CONVERGENCIA:** Repetir hasta que Q-values se estabilicen

**FUNCIÓN DE ACTUALIZACIÓN EXPLICADA:**
- **Q(s,a)**: Valor actual estimado para el par estado-acción
- **α**: Tasa de aprendizaje [0,1] - controla cuánto se actualiza el Q-value
- **r**: Recompensa inmediata observada
- **γ**: Factor de descuento [0,1] - importancia de recompensas futuras
- **max_a' Q(s',a')**: Mejor Q-value posible desde el siguiente estado s'
- **[r + γ max_a' Q(s',a') - Q(s,a)]**: Error TD (diferencia temporal)

El algoritmo es **OFF-POLICY** porque aprende Q* independientemente de la política seguida durante la exploración (ε-greedy).

### Justificación del Entorno CliffWalking-v0

**DESCRIPCIÓN DEL ENTORNO:**

CliffWalking-v0 es un entorno de navegación en grilla donde un agente debe moverse desde una posición inicial hasta un objetivo evitando caer por un acantilado.

**Características del Entorno:**
- **Espacio de Estados**: Discreto, 48 estados (grilla 4x12)
  - Estado inicial: posición (3,0) - esquina inferior izquierda
  - Estado objetivo: posición (3,11) - esquina inferior derecha  
  - Estados peligrosos: posiciones (3,1) a (3,10) - el acantilado

- **Espacio de Acciones**: Discreto, 4 acciones
  - 0: Arriba (Norte)
  - 1: Derecha (Este)
  - 2: Abajo (Sur)
  - 3: Izquierda (Oeste)

- **Sistema de Recompensas**:
  - -1: Por cada paso normal (incentiva encontrar camino corto)
  - -100: Por caer al acantilado (fuerte penalización)
  - 0: Por alcanzar el objetivo (terminación exitosa)

- **Terminación**: El episodio termina cuando el agente alcanza el objetivo o cae al acantilado

**JUSTIFICACIÓN DE IDONEIDAD PARA Q-LEARNING:**

1. **Espacios Discretos**: Q-Learning clásico requiere espacios de estados y acciones discretos y finitos para poder usar una Q-table. CliffWalking cumple perfectamente esta condición con 48 estados y 4 acciones.

2. **Tamaño Manejable**: Con 48×4 = 192 entradas en la Q-table, el problema es lo suficientemente pequeño para Q-Learning tabular pero no trivial.

3. **Balance Exploración/Explotación**: El entorno presenta un trade-off interesante:
   - **Camino Seguro**: Ir arriba, luego derecha, luego abajo (más pasos, recompensa ≈ -25)
   - **Camino Óptimo**: Ir directamente por el borde del acantilado (menos pasos, recompensa ≈ -13, pero riesgoso durante exploración)

4. **Recompensas Bien Diseñadas**: La estructura de recompensas (-1 por paso, -100 por caída) crea un problema de optimización claro que Q-Learning puede resolver.

5. **Propiedad de Markov**: El siguiente estado depende únicamente del estado actual y la acción tomada, cumpliendo la asunción fundamental de los MDPs.

### Justificación de Hiperparámetros Seleccionados

**1. Tasa de Aprendizaje (α = 0.1)**
- **Rango típico**: 0.01 - 0.5
- **Justificación**: 0.1 proporciona un balance entre:
  - Aprendizaje suficientemente rápido para converger en tiempo razonable
  - Estabilidad para evitar oscilaciones en los Q-values
  - Para CliffWalking, permite ajustes graduales sin sobrescribir bruscamente conocimiento previo

**2. Factor de Descuento (γ = 0.95)**
- **Rango típico**: 0.9 - 0.99
- **Justificación**:
  - Valor alto (cercano a 1) porque queremos que el agente considere recompensas futuras
  - En CliffWalking, las decisiones tempranas afectan significativamente el resultado final
  - 0.95 evita problemas numéricos que podrían surgir con γ = 0.99

**3. Epsilon Inicial (ε = 1.0)**
- **Justificación**: Comenzar con exploración completa es estándar en Q-Learning
- Permite al agente descubrir todas las transiciones posibles del entorno
- Especialmente importante en CliffWalking para descubrir tanto el camino seguro como el óptimo

**4. Decaimiento de Epsilon (decay = 0.995)**
- **Justificación**: Decaimiento lento y gradual
- Permite exploración extendida (crucial para encontrar el camino óptimo arriesgado)
- Asegura convergencia eventual hacia política greedy

**5. Epsilon Mínimo (ε_min = 0.01)**
- **Justificación**: Mantiene mínima exploración para evitar convergencia a políticas subóptimas
- 1% de exploración residual es suficiente para CliffWalking

**6. Número de Episodios (1000)**
- **Justificación**:
  - Suficiente para convergencia en entorno relativamente simple
  - Permite observar curva de aprendizaje completa
  - Balance entre tiempo computacional y calidad de aprendizaje

````python
import numpy as np
import gymnasium as gym
import matplotlib.pyplot as plt
from gymnasium.wrappers import TimeLimit
from collections import defaultdict
import random

class QLearningAgent:
    """
    Implementación del algoritmo Q-Learning desde cero.

    RELACIÓN CON TEORÍA MDP:
    - Mantiene Q-table que mapea pares (estado,acción) a valores Q
    - Implementa política ε-greedy para balance exploración/explotación
    - Usa ecuación de Bellman para actualización incremental de valores
    """

    def __init__(self, n_states, n_actions, learning_rate=0.1, discount_factor=0.95,
                 epsilon=1.0, epsilon_decay=0.995, epsilon_min=0.01):
        """
        INICIALIZACIÓN según teoría Q-Learning:

        Args:
            n_states: |S| - Cardinalidad del espacio de estados del MDP
            n_actions: |A| - Cardinalidad del espacio de acciones del MDP
            learning_rate (α): Tasa de aprendizaje - controla velocidad de actualización
            discount_factor (γ): Factor de descuento - importancia recompensas futuras
            epsilon (ε): Probabilidad de exploración en política ε-greedy
        """
        self.n_states = n_states
        self.n_actions = n_actions
        self.alpha = learning_rate           # α en ecuación de actualización
        self.gamma = discount_factor         # γ en ecuación de actualización
        self.epsilon = epsilon               # ε para política ε-greedy
        self.epsilon_decay = epsilon_decay   # Decaimiento de exploración
        self.epsilon_min = epsilon_min       # ε mínimo para mantener exploración

         # PASO 1 TEORÍA: Inicializar Q-table con ceros para todos los pares (s,a)
        self.q_table = np.zeros((n_states, n_actions))

    def choose_action(self, state):
        """
        IMPLEMENTA POLÍTICA ε-GREEDY descrita en teoría:

        - Con probabilidad ε: EXPLORACIÓN (acción aleatoria)
        - Con probabilidad 1-ε: EXPLOTACIÓN (acción óptima según Q-table actual)

        Esta política garantiza que el agente explore suficientemente el espacio
        estado-acción mientras explota el conocimiento adquirido.
        """
        if random.uniform(0, 1) < self.epsilon:
            # EXPLORACIÓN: Selección uniforme aleatoria del espacio de acciones A
            return random.choice(range(self.n_actions))
        else:
            # EXPLOTACIÓN: argmax_a Q(s,a) - política greedy respecto a Q-table
            return np.argmax(self.q_table[state])

    def update_q_value(self, state, action, reward, next_state, done):
        """
        IMPLEMENTA ECUACIÓN DE ACTUALIZACIÓN DE BELLMAN para Q-Learning:
        Q(s,a) ← Q(s,a) + α[r + γ max_a' Q(s',a') - Q(s,a)]

        COMPONENTES TEÓRICOS:
        - Q(s,a): Estimación actual del valor de tomar acción a en estado s
        - α: Tasa de aprendizaje (paso de descenso de gradiente)
        - r: Recompensa inmediata R(s,a,s') del MDP
        - γ: Factor de descuento para recompensas futuras
        - max_a' Q(s',a'): Valor óptimo estimado desde siguiente estado (BOOTSTRAP)
        - Error TD: Diferencia entre estimación actual y objetivo mejorado
        """
        # Obtener estimación actual Q(s,a)
        current_q = self.q_table[state, action]

        # CALCULAR OBJETIVO (TARGET) según ecuación de Bellman
        if done:
            # Estado terminal: V*(s') = 0, solo recompensa inmediata
            target = reward
        else:
            # Estado no terminal: r + γ max_a' Q(s',a')
            # BOOTSTRAP: Usar estimación actual de Q para aproximar V*(s')
            max_next_q = np.max(self.q_table[next_state])
            target = reward + self.gamma * max_next_q

        # ERROR TD (Temporal Difference): diferencia entre objetivo y estimación actual
        td_error = target - current_q

        # ACTUALIZACIÓN INCREMENTAL: Q(s,a) ← Q(s,a) + α * TD_error
        # Esto implementa descenso de gradiente estocástico sobre función de pérdida
        self.q_table[state, action] = current_q + self.alpha * td_error

    def decay_epsilon(self):
        """
        IMPLEMENTA ESTRATEGIA DE DECAIMIENTO DE EXPLORACIÓN:

        Reduce ε gradualmente para transicionar de exploración (inicio) a
        explotación (final del entrenamiento). Esto es crucial para convergencia
        ya que garantiza que el agente eventualmente siga política greedy óptima.
        """
        self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay)

    def get_policy(self):
        """
        EXTRAE POLÍTICA DETERMINISTA GREEDY de la Q-table aprendida:
        π*(s) = argmax_a Q*(s,a)

        Esta política representa la estrategia óptima aprendida por el agente.
        """
        return np.argmax(self.q_table, axis=1)

def train_q_learning(env, agent, n_episodes=1000, max_steps_per_episode=500):
    """
    IMPLEMENTA BUCLE PRINCIPAL DE ENTRENAMIENTO Q-Learning según teoría:

    1. Para cada episodio:
       a) Reiniciar entorno (estado inicial del MDP)
       b) Ejecutar pasos hasta estado terminal o límite de tiempo
       c) En cada paso: observar, actuar, aprender (ciclo agente-entorno)

    2. Recopilar métricas de aprendizaje para análisis de convergencia
    """
    rewards_per_episode = []
    steps_per_episode = []

    for episode in range(n_episodes):
        # REINICIAR EPISODIO: s_0 ~ μ (distribución inicial de estados)
        state, _ = env.reset()
        total_reward = 0
        steps = 0

        for step in range(max_steps_per_episode):
            # CICLO AGENTE-ENTORNO descrito en teoría:

            # 1. AGENTE SELECCIONA ACCIÓN según política ε-greedy
            action = agent.choose_action(state)

            # 2. ENTORNO RESPONDE con nueva observación y recompensa
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated

            # 3. AGENTE APRENDE de la experiencia (s,a,r,s')
            # Implementa actualización incremental de Q-Learning
            agent.update_q_value(state, action, reward, next_state, done)

            # 4. TRANSICIÓN DE ESTADO para siguiente iteración
            state = next_state
            total_reward += reward
            steps += 1

            # CONDICIÓN DE PARADA: Estado terminal o límite temporal
            if done:
                break

        # ESTRATEGIA DE EXPLORACIÓN: Reducir ε después de cada episodio
        # Implementa schedule de exploración decreciente para convergencia
        agent.decay_epsilon()

        # MÉTRICAS DE ENTRENAMIENTO para análisis posterior
        rewards_per_episode.append(total_reward)
        steps_per_episode.append(steps)

        # MONITOREO DE PROGRESO cada 100 episodios
        if (episode + 1) % 100 == 0:
            avg_reward = np.mean(rewards_per_episode[-100:])
            print(f"Episodio {episode + 1}, Recompensa promedio (últimos 100): {avg_reward:.2f}, ε: {agent.epsilon:.3f}")

    return rewards_per_episode, steps_per_episode

def test_policy(env, agent, n_test_episodes=100):
    """
    Evalúa la política aprendida sin exploración (ε=0).

    Args:
        env: Entorno de prueba
        agent: Agente entrenado
        n_test_episodes: Número de episodios de prueba

    Returns:
        success_rate: Porcentaje de episodios completados exitosamente
        avg_reward: Recompensa promedio
        avg_steps: Número promedio de pasos
    """
    successful_episodes = 0
    total_rewards = []
    total_steps = []

    # Guardar epsilon original y establecer a 0 (sin exploración)
    original_epsilon = agent.epsilon
    agent.epsilon = 0.0

    for episode in range(n_test_episodes):
        state, _ = env.reset()
        total_reward = 0
        steps = 0

        while True:
            action = agent.choose_action(state)
            state, reward, terminated, truncated, _ = env.step(action)
            total_reward += reward
            steps += 1

            if terminated or truncated:
                if terminated and reward > 0:  # Éxito si termina con recompensa positiva
                    successful_episodes += 1
                break

        total_rewards.append(total_reward)
        total_steps.append(steps)

    # Restaurar epsilon original
    agent.epsilon = original_epsilon

    success_rate = (successful_episodes / n_test_episodes) * 100
    avg_reward = np.mean(total_rewards)
    avg_steps = np.mean(total_steps)

    return success_rate, avg_reward, avg_steps

# Configuración e implementación en el entorno CliffWalking-v0
print("=== IMPLEMENTACIÓN DE Q-LEARNING ===")

# Crear entorno con TimeLimit wrapper
env = gym.make('CliffWalking-v0')
env = TimeLimit(env, max_episode_steps=500)

print(f"Entorno: CliffWalking-v0")
print(f"Espacio de estados: {env.observation_space.n}")
print(f"Espacio de acciones: {env.action_space.n}")

# Inicializar agente Q-Learning
agent = QLearningAgent(
    n_states=env.observation_space.n,
    n_actions=env.action_space.n,
    learning_rate=0.1,      # α
    discount_factor=0.95,   # γ
    epsilon=1.0,            # ε inicial
    epsilon_decay=0.995,    # Decaimiento de ε
    epsilon_min=0.01        # ε mínimo
)

print(f"\nHiperparámetros:")
print(f"- Tasa de aprendizaje (α): {agent.alpha}")
print(f"- Factor de descuento (γ): {agent.gamma}")
print(f"- Epsilon inicial (ε): {agent.epsilon}")
print(f"- Decaimiento de epsilon: {agent.epsilon_decay}")
print(f"- Epsilon mínimo: {agent.epsilon_min}")

# Entrenamiento
print(f"\n=== ENTRENAMIENTO ===")
n_training_episodes = 1000
rewards, steps = train_q_learning(env, agent, n_episodes=n_training_episodes)

# Evaluación
print(f"\n=== EVALUACIÓN ===")
success_rate, avg_reward, avg_steps = test_policy(env, agent, n_test_episodes=100)

print(f"Resultados después de {n_training_episodes} episodios de entrenamiento:")
print(f"- Tasa de éxito: {success_rate:.1f}%")
print(f"- Recompensa promedio: {avg_reward:.2f}")
print(f"- Pasos promedio: {avg_steps:.1f}")

# Visualización de resultados
plt.figure(figsize=(15, 5))

# Gráfico 1: Recompensas por episodio
plt.subplot(1, 3, 1)
plt.plot(rewards)
plt.title('Recompensas por Episodio')
plt.xlabel('Episodio')
plt.ylabel('Recompensa Total')
plt.grid(True)

# Gráfico 2: Promedio móvil de recompensas
plt.subplot(1, 3, 2)
window_size = 50
moving_avg = [np.mean(rewards[max(0, i-window_size):i+1]) for i in range(len(rewards))]
plt.plot(moving_avg)
plt.title(f'Promedio Móvil de Recompensas (ventana={window_size})')
plt.xlabel('Episodio')
plt.ylabel('Recompensa Promedio')
plt.grid(True)

# Gráfico 3: Pasos por episodio
plt.subplot(1, 3, 3)
plt.plot(steps)
plt.title('Pasos por Episodio')
plt.xlabel('Episodio')
plt.ylabel('Número de Pasos')
plt.grid(True)

plt.tight_layout()
plt.show()

# Mostrar parte de la Q-table aprendida
print(f"\n=== Q-TABLE APRENDIDA (primeros 10 estados) ===")
print("Estado | Arriba | Derecha | Abajo | Izquierda")
print("-" * 50)
for state in range(min(10, agent.n_states)):
    q_values = agent.q_table[state]
    print(f"{state:6d} | {q_values[0]:6.2f} | {q_values[1]:7.2f} | {q_values[2]:5.2f} | {q_values[3]:9.2f}")

# Política aprendida
policy = agent.get_policy()
action_names = ['Arriba', 'Derecha', 'Abajo', 'Izquierda']
print(f"\n=== POLÍTICA APRENDIDA (primeros 10 estados) ===")
for state in range(min(10, agent.n_states)):
    best_action = policy[state]
    print(f"Estado {state}: {action_names[best_action]}")
````

**Aprendizaje por Refuerzo Profundo**
El aprendizaje por Refuerzo Profundo (DRL por sus siglas en inglés) surge de la unión del aprendizaje por refuerzo con el aprendizaje profundo. La principal diferencia respecto al aprendizaje por refuerzo es que este suele usar tablas para almacenar los valores de los estados o pares estado-acción, lo que limita mucho la escalabilidad, mientras que el DRL utilizar **redes neuronales profundas** para aproximar estas funciones de valor o directamente las políticas.

El principal objetivo de esta rama es superar el problema conocido en el ámbito como la "maldición de la dimensionalidad". Este problema lo vemos principalmente en entrnos como los videojuegos o la robótica en los que el nº de estados posibles es astronómico. Al usar redes neuronales, el DRL puede aprender representaciones complejas de estos estados y acciones, generalizando el conocimiento a estados no vistos y permitiendo manejar entornos con alta dimensionalidad. Esto ha llevado a avances significativos, como el famoso ejemplo de superación del ser humano en el Go.

**Categorías básicas derivadas de depp Q-Networks (DQN) y Policy Gradient**

En DRL, los algoritmos se pueden clasificar en dos categorías principales basadas en la aproximación que realizan con las redes neuronales:
* **Métodos Basados en Valores - Derivados de DQN**:
  1. Fundamento: estos algoritmos utilizan una red neuronal para aproximar la Q(s,a). Es decir, la red toma un estado como entrada y produce como salida los valores Q estimados para cada acción posible en ese estado, buscando aprender Q*(s,a) que representa el valor óptimo de tomar una acción a en un estado s.
  2. Funcionamiento: una vez que la Q-función es aproximada por la red, se deriva la política de forma greedy, es decir, en cada estado el agente selecciona la acción que tiene el Q-valor más alto según la red neuronal.
  3. Aprendizaje: la red ajusta sus pesos para que los Q-valores predicados se acerquen a los valores objetivos calculados a partir de las recompensas observadas y los Q-valores de los siguientes estados.
  4. Ventajas: suelen ser más estables y eficientes en términos de muestreo que los métodos basados en política pura, sobre todo en entornos con espacios de acción discretos.
  5. Ejemplos: algunos ejemplos son DQN, Dueling DQN y Prioritized Experience Replay.
* **Métodos Basados en Políticas - Derivados de Policy Gradient**
  1. Fundamento: estos algoritmos usan una red neuronal para aproximar directamente la política del agente. LA red toma un estado como entrada y produce como salida una distribución de probabilidad sobre las acciones posibles. Tras esto, el agente muestrea una acción de esta distribución.
  2. Funcionamiento: no se busca aprender el valor de los estados o acciones, sino directamente aprender una función que mapee estados a acciones que maximicen la recompensa.
  3. Aprendizaje: la red se entrena directamente para optimizar una función objetivo que mide la recompensa esperada. Esto se realiza ajustando los pesos de la red en la dirección del gradiente de la política (de ahí viene Policy Gradient), esto quiere decir que las acciones que conducen a recompensas más altas se vuelven más probables.
  4. Ventajas: pueden aprender políticas aleatorias, lo que viene bien para la exploración y en entornos donde una política determinista podría no ser óptima. Sobre todo se le saca provecho en espacios de acción continuos.
  5. Desventajas: pueden tener una convergencia más lenta y una mayor varianza en las estimaciones del gradiente.
  6. Ejemplos: REINFORCE, Actor-Critic(A2C, A3C, DDPG...) y PPO.

**Algoritmo seleccionado: Deep Q-Network (DQN)**

Hemos seleccionado este algoritmo para ñla descripción detallada ya que fue uno de los primeros y más influyentes de DRL, demostrando un gran éxito al aprender a jugar juegos de Atari, superando a los humanos en muchos de ellos. Además como se indica en el enunciado, es uno de los tradicionales.

**Funcionamiento General de DQN**

DQN extiende el algoritmo clásico Q-Learning utilizando una red neuronal profunda para aproximar la función Q(s,a). El principal cambio es que en vez de usar una Q-Table, usa una Q-Network, que toma el estado del entorno como entrada y produce un Q-valor para cada acción posible como salida.
El entrenamiento de DQN se basa en la idea de minimizar el error de diferencia temporal (TD error), parecido a Q-Learning, pero con dos innovaciones clave para estabilizar el entrenamiento con redes neuronales:
1. **Experience Replay**: las transiciones (estado actual, acción, recompensa, siguiente estado) se almacenan en un búfer de repetición (replay buffer). A lo largo del entrenamiento, se muestrean aleatoriamente pequeños batches de estas transiciones del búfer para actualizar la red. Esto rompe la correlación entre muestras secuenciales, lo que ayuda a estabilizar el entrenamiento de la red neuronal.
2. **Target Network**: se usa una segunda red neuronal conocida como target network que es una copia retrasada de la Q-Network principal, cuyo objetivo principal es calcular los Q-valores objetivos en la ecuación. Actualizar la red objetivo de forma infrecuente (copiando los pesos de la red principal cada cierto número de pasos o de forma suave) estabiliza el objetivo de entrenamiento, evitando oscilaciones y divergencias que podrían ocurrir si la misma red se usara tanto para predecir como para calcular el objetivo.

**Componentes clave de DQN**

1. Agente: es el cerebro del sistema, contiene la Q-Network y la lógica para seleccionar acciones y aprender.
2. Entorno: El mundo con el que el agente interactúa, proporcionando estados, recompensas y transiciones.
3. Q-Network: una red neuronal profunda con pesos θ. Toma un estado s como entrada y produce un vector de Q-valores, uno para cada acción a en ese estado: Q(s,a;θ).
4. Target Network: una copia de la Q-Network con pesos θ′. Se usa para calcular los valores objetivo y se actualiza periódicamente copiando los pesos de la Q-Network principal.
5. Replay Buffer: un almacén de transiciones (s, a, r, s', done), en el que done indica si s' es un estado terminal.
6. Política de exploración (ϵ-greedy): durante el entrenamiento, el agente elige acciones usando una política ϵ-greedy: con probabilidad ϵ elige una acción aleatoria (exploración), y con probabilidad 1−ϵ elige la acción con el Q-valor más alto predicho por la Q-Network (explotación). ϵ suele decrecer con el tiempo.

**Funcionamiento y entrenamiento de la Q-Network**

Este se puede resumir en los siguientes pasos:
1. **Inicialización**
  * Inicializar la Q-Network principal con pesos aleatorios θ.
  * Inicializar la Target Network con los mismos pesos θ′=θ.
  * Inicializar el Replay Buffer D vacío.
  * Definir hiperparámetros: tasa de aprendizaje (α), factor de descuento (γ), tamaño del mini-batch, capacidad del replay buffer, frecuencia de actualización de la target network (C), ϵ inicial y final, y número de episodios/pasos.
2. **Bucle para cada episodio**
  * Reiniciar el entorno y obtener el estado inicial s.
  * Bucle para cada paso del episodio:
    * Selección de acción (a): Con probabilidad ϵ, elige una acción aleatoria. Con probabilidad 1-ϵ, elige la acción a*=argmax Q(s,a;θ) (la acción con el Q-value más alto predicho por la Q-Network principal para el estado actual s).
    * Ejecutar acción: Realiza la acción a en el entorno.
    * Observar retroalimentación: el entorno devuelve la recompensa r, el nuevo estado s' y una señal done (indica si s' es estado terminal)
    * Almacenar transición: guarda la tupla (s, a, r, s', done) en el Replay Buffer D.
    * Actualizar estado: s <- s'.
    * Condición de fin de episodio: si done es verdadero, terminar el episodio.
    * **Entrenamiento de la Q-Network**:
      * Generación del conjunto de datos de entrada y salida (Mini-batch):
        * Muestrear aleatoriamente un mini-batch de transiciones del Replay Buffer D. Estas serán las entradas para el entrenamiento.
        * Para cada transición en el mini-batch, calcular el valor objetivo y:
          * Si s' es un estado terminal, y=r (solo se considera la recompensa inmediata, ya que no hay recompensas futuras desde un estado terminal).
          * Si s no es terminal, y = r + γ max Q(s', a', θ') (la recompensa inmediata más el Q-value máximo descontado de siguiente estado s', que se calcula usando la Target Network con pesos θ').
    * **Cálculo de la Función de Pérdida**:
      La más común es el Error Cuadrático Medio (MSE), que mide la diferencia entre los Q-valores predichos por la red principal y los valores objetivos calculados:
      L(θ)= 1/N N∑j=1 (y - Q(s, a, θ))^2
      Donde:
            N: Tamaño del mini-batch
            y: el valor objetivo calculado para esta transición.
            Q(s, a, θ): el Q-value predicho por la Q-Network principal para la acción a tomada ne el estado s.
    * **Optimización y Actualización de Pesos**
      * Se realiza un paso de descenso de gradiente (con Adam, SGD...) para actualizar los pesos θ de la Q-Network principal, minimizando la función de perdida L(θ). Esti se consigue calculando el gradiente de la pérdida con respecto a los pesos θ y moviendo los pesos en la dirección opuesta al gradiente.
    * **Actualización de la Target Network**
      * Cada C pasos, se copian los pesos actuales de la Q-Network pirncipal a la Target Network (θ' <- θ). Así se asgura que la red objetivo se mantenga actualizada, pero con un retraso para la estabilidad.
  * **Decaimiento de ϵ**: se decrementa el valor de ϵ gradualmente a lo largo del entrenamiento para reducir la exploración y aumentar la explotación a medida que el agente aprende.
3. **Convergencia**: el entrenamiento continua hasta que el agente consigue un rendimiento satisfactorio o se alcance un nº máximo de episodios.

A modo de resumen:
1. **Red de Política**: En DQN, la "red de política" no es una red separada que emita probabilidades de acciones. En su lugar, la política es implícitamente definida por la Q-Network: la política es siempre determinista (greedy) después del entrenamiento, eligiendo la acción con el Q-valor más alto. Durante el entrenamiento, la política de exploración (ϵ-greedy) se usa para generar experiencia.
2. **Entrenamiento de la Red**: La Q-Network se entrena como cualquier otra red neuronal, utilizando datos (mini-batches de transiciones del replay buffer), una función de pérdida (MSE) y un optimizador (descenso de gradiente).
3. **Generación de Datos**: Los datos se generan a través de la interacción del agente con el entorno y se almacenan en el replay buffer. La selección de un mini-batch aleatorio de este buffer constituye el "conjunto de datos" para cada paso de entrenamiento.
4. **Función de Pérdida**: El MSE se utiliza para medir la diferencia entre lo que la red predice actualmente y lo que debería predecir según la Ecuación de Bellman (los valores objetivo).
5. **Componentes**: La Q-Network principal y la Target Network son las "redes" del algoritmo. El replay buffer es crucial para descorrelacionar los datos y estabilizar el entrenamiento.
